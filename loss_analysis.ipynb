```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
for epoch in range(50):
    filename = 'loss_epoch_{:04d}'.format(epoch+1)
    df = pd.concat([df, pd.read_csv(filename)], ignore_index=True)
```

```python
plt.plot(df['gen'], label='Generator')
plt.plot(df['disc'], label='Discriminator')
plt.legend()
plt.show()
```

```python
pd.set_option('display.max_rows', 500)
df
```

```python
gen_losses, disc_losses = [], []
for epoch in range(50):
    filename = 'loss_epoch_{:04d}'.format(epoch+1)
    means = pd.read_csv(filename).mean(axis=0)
    gen_losses.append(means['gen'])
    disc_losses.append(means['disc'])
```

```python
plt.plot(gen_losses, label='Generator')
plt.plot(disc_losses, label='Discriminator')
plt.legend()
plt.show()
```
